```python
# 행렬을 쌓는 함수
# hstack = 가로로
# vstack = 세로로

x = np.arange(10)
y = 2*x

print(x.shape)
print(y.shape)

# numpy 벡터 = 행 벡터가 디폴트
z1 = np.hstack([x, y])
z2 = np.vstack([x, y])

print(z1.shape) # 벡터 ndim = 1짜리인, shape = (20, )
print(z2.shape) # 1 x 10짜리 벡터 두개를, 세로로 쌓은듯한, 2 x 10짜리 행렬
```

```python
X = np.random.rand(10, 2)
print(X)

print(np.sum(X)) # 전체 값을 다 더함, 스칼라 한개
print(np.sum(X, axis = 0)) # 더하는 방향을 0번째 축, 행에 대해서 덧셈을 해달라, 결과는 열
print(np.sum(X, axis = 1)) # 더하는 방향을 1번째 축, 열에 대해서 덧셈을 해달라, 결과는 행

print(np.mean(X)) # 전체 값에 대한 평균
print(np.mean(X, axis = 0)) # 행에 대해서 평균
print(np.mean(X, axis = 1)) # 열에 대해서 평균
```

```python
x = np.random.rand(10)

print(x)
print(np.max(x)) # 들어있는 값들 중에 제일 큰 값을 리턴
print(np.argmax(x)) # 들어있는 값들 중에 제일 큰값이 있는 인덱스를 리턴
print(np.min(x))
print(np.argmin(x))

# np.max(x) == x[np.argmax(x)]
# np.min(x) == x[np.argmin(x)]

print(np.where(x >= 0.5)) # 인덱스를 리턴
```
